## 手順
　まずはNumPyとBokehをインポートする。

````python
import numpy as np
import bokeh.plotting as bkh
bkh.output_notebook()
````

### Take1
　ランダムデータを作成し、線グラフを描画する。対話型グラフがnotebookに描画され、グラフ上にあるボタンで、パンやズームが可能。

````python
x = np.linspace(0., 1., 100) # グラフ作成用に、NumPy配列のランダムデータを作成
y = np.cumsum(np.random.randn(100))
````

````python
#bkh.line(x, y, line_width=5)
#bkh.show()

p = bkh.figure()
p.line(x=x, y=y)
bkh.show(p)
````

### Take2
　次にサンプルとして提供されている(アイリス)データセットをロードし、対話型の散布図を描画する。

````python
from bokeh.sampledata.iris import flowers
colormap = {'setosa': 'red', 'versicolor':  'green', 'virginica': 'blue'} # 花の色に基づいた配色の設定
flowers['color'] = flowers['species'].map(lambda x: colormap[x])
````

````python
p = bkh.figure()
p.scatter(flowers["petal_length"], 
          flowers["petal_width"],
          color=flowers["color"], 
          fill_alpha=0.25, size=10,)
bkh.show(p)
````
