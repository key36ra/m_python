```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 

```python
g_x = lambda t: t*np.cos(t)
g_y = lambda t: t*np.sin(t)
```

```python
t = np.linspace(0., 2 * np.pi, 400)
```

```python
x_tr = g_x(t)
y_tr = g_y(t)
```

```python
plt.plot(x_tr[:200], y_tr[:200], 'k')
```

\begin{eqnarray}
L &=& asin(\theta)+b \\
x &=& Lcos(\theta) = asin(\theta)cos(\theta) \\
y &=& Lsin(\theta) = asin(\theta)sin(\theta) \\
\end{eqnarray}

```python
f_x = lambda t: (np.sin(t)+1)*np.cos(t)
f_y = lambda t: (np.sin(t)+1)*np.sin(t)
```

```python
x = f_x(t)
y = f_y(t)
```

```python
plt.plot(x[:400], y[:400], 'k')
plt.xlabel('x')
plt.ylabel('y')
```
