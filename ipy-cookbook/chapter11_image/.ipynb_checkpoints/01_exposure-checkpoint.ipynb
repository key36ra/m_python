# 11.1 画像の露出補正
　画像の露出を見れば、画像が明るいのか、暗いのか、適正なのかを判断できる。画像の明るさはヒストグラムで把握できる。露出の改善は、画像の露出の基礎的な操作であり、scikit-imageを使えば簡単にできる。

## 準備

 ```python
 conda install scikit-image
 ```
 をターミナルから実行する。
 
　また、本書の[GitHubリポジトリ](https://github.com/ipython-books/cookbook-data)からbeach.zipをダウンロードし、展開する必要がある。

## 手順
1. 必要なパッケージのインストール。

````python
import numpy as np
import matplotlib.pyplot as plt
import skimage.exposure as skie
%matplotlib inline
````

1. 画像をmatplotlibで読み込む。グレイスケール画像として、1つのRGB要素だけを使う。

````python
img = plt.imread('data/pic1.jpg')[...,0]
````

1. 画像と、画像のヒストグラム(明るさの分布グラフで露出を表す)を併せて表示する関数を作る。

````python
def show(img):
    # 画像を表示
    plt.subplot(121)
    plt.imshow(img, cmap=plt.cm.gray)
    plt.axis('off')
    # ヒストグラムを表示
    plt.subplot(122)
    plt.hist(img.ravel(), lw=0, bins=256)
    plt.xlim(0, img.max())
    plt.yticks([])
    plt.show()
````

1. 画像とヒストグラムを表示する。

````python
show(img)
````

　ヒストグラムのバランスが悪く、画像は露出オーバー(多くの画素が明るすぎる)気味である。
1. scikit-imageのrescale_intensity関数を使って、画素の明るさを補正する。in_rangeとout_range引数は、元画像と習性画像の線形写像を指定する。in_rangeの範囲外の明るさをモツ画素はout_rangeの範囲に補正される。つまり、暗い(明るさが100以下)画素は、真黒(0)に、明るい(明るさが240を超える)画素は真白になる。

````python
show(skie.rescale_intensity(img, in_range=(100, 240), out_range=(0, 255)))
````

1. 次に、より高度な露出補正手法である**コントラスト制限付き適応ヒストグラム均等化**(Contrast Limited Adaptive Histogram Equalization : CLAHE)を使う。

````python
show(skie.equalize_adapthist(img))
````
