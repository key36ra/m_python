# 8.1 はじめてのscikit-learn
　ここでは、[scikit-learn](http://scikit-learn.org)機械学習パッケージの基礎を紹介する。このパッケージは、本書での機械学習を通して使う主要なツール。明快なAPIにより、モデルの定義、訓練およびテストが容易に行える。加えて、処理速度向上のためにデザインされており、ビッグデータ向けである。

　ここでは、非常に基本的な線形回帰の例として曲線のフィッティングを紹介する。この簡単な例により、線形モデル、過学習、未学習、正則化、交差検証などの鍵となる概念が学べる。

 を実行する。

## 手順
　簡単なモデルに従う(ノイズを含む)1次元のデータを生成し、データに対する関数のあてはめを行う。その上でこの関数を用いて、異なるデータポイントの値を予測する。これは**曲線フィッティング回帰問題**と呼ばれる。
1. まずは必要なパッケージをインポートする。

````python
import numpy as np
import scipy.stats as st
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
f = lambda x: np.exp(3 * x) # 入力xのexpを返す関数f
````

````python
x_tr = np.linspace(0., 2, 200) # x軸区間[0,2]で200点
y_tr = f(x_tr) # デジタルx値からexp関数生成
````

````python
x = np.array([0, .1, .2, .5, .8, .9, 1]) # リストから配列を生成
y = f(x) + np.random.randn(len(x)) # xを関数に代入し、ノイズを乗せる
````

````python
plt.plot(x_tr[:100], y_tr[:100], '--k')
plt.plot(x, y, 'ok', ms=10)
````

````python
lr = lm.LinearRegression() # モデルを作成
lr.fit(x[:, np.newaxis], y) # 訓練データを使って学習を行う
y_lr = lr.predict(x_tr[:, np.newaxis]) # 学習済みモデルで予測を行う
````

````python
plt.plot(x_tr, y_tr, '--k')
plt.plot(x_tr, y_lr, 'g')
plt.plot(x, y, 'ok', ms=10)
plt.xlim(0, 1)
plt.ylim(y.min()-1, y.max()+1)
plt.title("Linear regression")
````

````python
lrp = lm.LinearRegression()
plt.plot(x_tr, y_tr, '--k')
for deg in [2, 5]:
    lrp.fit(np.vander(x, deg + 1), y)
    y_lrp = lrp.predict(np.vander(x_tr, deg + 1))
    plt.plot(x_tr, y_lrp, label='degree ' + str(deg))
    plt.legend(loc=2)
    plt.xlim(0, 1.4)
    plt.ylim(-10, 40)
    # モデルの係数を表示
    print(' '.join(['%.2f' % c for c in lrp.coef_]))
plt.plot(x, y, 'ok', ms=10)
plt.title("Linear regression")
````

````python
ridge = lm.RidgeCV()
plt.plot(x_tr, y_tr, '--k')

for deg in [2, 5]:
    ridge.fit(np.vander(x, deg + 1), y)
    y_ridge = ridge.predict(np.vander(x_tr, deg+1))
    plt.plot(x_tr, y_ridge, label='degree ' + str(deg))
    plt.legend(loc=2)
    plt.xlim(0, 1.5)
    plt.ylim(-5, 80)
    # モデルの係数を表示
    print(' '.join(['%.2f' % c for c in ridge.coef_]))
plt.plot(x, y, 'ok', ms=10)
plt.title("Ringe regression") # リッジ回帰
````
