# 13.4 確率微分方程式のシミュレーション
　**確率微分方程式**(Stochastic differential equations : SDE)は、ノイズを含む力学系をモデル化する。  
　このレシピでは、**ランジュバン方程式**の解である**オルシュタイン・ウーレンベック過程**をシミュレーションする。このモデルは、抵抗のある流体内における粒子の確率的変化を表す。流体分子の衝突(拡散)により粒子に動きが生じる。定稿の影響下にある点がブラウン運動との相違である。  
　オルシュタイン・ウーレンベック過程は、定常的でガウス性とマルコフ聖を満たす定常ランダムノイズの良い候補である。  
　ここでは**オイラー・丸山法**と呼ばれる数値解法を使ってシミュレーションする。これは常微分方程式に対するオイラー法を確率微分方程式へと単純に一般化したものである。

## 手順
1. NumPyとmatplotlibをインポート。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

1. モデルに対するパラメータを定義する。

```python
sigma = 1. # 標準偏差
mu = 10.0 # 平均
tau = 0.05 # 時定数
```

1. シミュレーションに対するパラメータを定義する。

```python
dt = 0.001 # 時間ステップ
T = 1.0 # 総時間
n = int(T/dt) # 時間ステップ数
t = np.linspace(0., T, n) # 時間ベクトル
```

1. (各時間ステップで、これらの定数を再計算しなくても良いように)次の関数を定義する。

```python
sigma_bis = sigma * np.sqrt(2. / tau)
sqrtdt = np.sqrt(dt)
```

1. シミュレーションの結果を格納する配列を作成する。

```python
x = np.zeros(n)
```

1. オイラー・丸山法を使ってシミュレートする。ODEに対する標準オイラー法に似ているが、(通常の確率変数である)確率項が追加されている。

```python
for i in range(n-1):
    x[i+1] = x[i] + dt*(-(x[i]-mu)/tau) + sigma_bis * sqrtdt * np.random.randn()
```

1. シミュレーション結果を表示する。

```python
plt.plot(t, x)
```

1. この確率過程の分散がどのように時間変化するか確認する。そのためには、ベクトル化を用いて同じ確率過程の独立したシミュレーションを何度も行う。配列Xを使って、ある時点におけるすべてのシミュレーションの値を保持する(言い換えると、すべてのシミュレーションのすべての時間における値をメモリ上に持つわけではない)。この配列は各時間ステップごとに上書きされる。

```python
ntrials = 10000
X = np.zeros(ntrials)
```

```python
# ヒストグラムのビンを作成する
bins = np.linspace(-2., 14., 50)
for i in range(n):
    # 各過程をそれぞれ更新する
    X += dt*(-(X-mu)/tau) + sigma_bis*sqrtdt*np.random.randn(ntrials)
    # 時間内のいくつかのてんでヒストグラムを描画する
    if i in (5, 50, 900):
        hist, _ = np.histogram(X, bins=bins)
        plt.plot((bins[1:]+bins[:-1])/2, hist, label="t={0:.2f}".format(i*dt))
    plt.legend()
```

この確率過程の分散は、平均$\mu$=10、標準偏差$\sigma$=1のガウス分布に近づく。適正なパラメータで初期の分散がガウス分布に従うならば、確率過程は時間に依存しない定常的となる。

## 解説
　このレシピで使用したランジュバン方程式は、次の確率微分方程式で表される。  
$$ dx = - \frac{(x-\mu)}{\tau} dt + \sigma \sqrt{\frac{2}{\tau}} dW $$
　ここで$x(t)$は確率過程、$dx$は微小量の増分、$\mu$は標準偏差、$\tau$は時間定数である。また、$W$はこのSDEの根底にあるブラウン運動(またはウィーナ過程)である。  
　右辺の第１項は、($dt$の)決定論的項であり、第2項は確率論的項である。この2項目がなければ、この方程式は通常の決定論的ODEと等しくなる。  
　ブラウン運動の微小ステップは、ガウス確率変数になる。特に(ある意味において)ブラウン運動の導関数は**ホワイトノイズ**つまり、独立ガウス確率変数列である。  
　オイラー・丸山法では、時間による離散化を行うとともに、各時間ステップに極小のステップを加える。この手法は、決定論的項(ODEに対する標準的なオイラー法)と確率論的項(ガウス確率変数)を含む。特に次の方程式  
$$ dx = a(t, x) dt + b(t,x) dW $$
に対する($t=n*dt$)の数値解析手法は次の式となる。  
$$ x_{n+1} = x_n + dx = x_n + a(t, x_n) dt + b(t, x_n) \sqrt{dt} \xi, (\xi \sim N(0,1)) $$
　ここで$\xi$は(各時間ステップで独立な)分散１のガウス確率変数である。正規化係数が$\sqrt{dt}$であるのは、ブラウン運動の微小ステップが標準偏差$\sqrt{dt}$を持つ事実から導かれている。
