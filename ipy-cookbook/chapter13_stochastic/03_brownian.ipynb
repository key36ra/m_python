# 13.3 ブラウン運動のシミュレーション
- **ブラウン運動**(または**ウィーナ過程**)は数学、物理学、その他多くの科学や工学分野における基礎的な研究対象である。このモデルは、流体中の分子が不規則に衝突することにより、流体中に浮遊する微粒子が不規則に運動する現象を説明する。より一般的に言うならば、微粒子が空間をあらゆる方向に独立かつ不規則に移動する連続時間**ランダムウォーク**をモデル化したものがブラウン運動である。
- 数学的には、ブラウン運動は連続時間マルコフ確率連鎖の一種。また、ブラウン運動は確率解析学、確率過程理論など数学の中心的な分野であるだけでなく、数理ファイナンス、生態学、神経科学など応用科学分野の重要な役割を果たしている。
- このレシピでは、2次元のブラウン運動をシミュレートし、その結果を可視化する。

## 手順
- まずNumPyとmatplotlibをインポートする。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

- 5000時間ステップのシミュレーションを行う。

```python
n = 5000
```

* 2つの独立した1次元ブラウン運動を組み合わせて、2次元のブラウン運動を構成する。(離散)ブラウン運動は各時間ステップで独立したガウシアンジャンプを行う。そのため、単純な乱数の累積値を計算し、各時間ステップの値とする。

```python
x = np.cumsum(np.random.randn(n))
y = np.cumsum(np.random.randn(n))
```

- このブラウン運動を表示する。plot(x, y)を使うだけでもよいが、そのままでは白黒の表示となるため時間の進行を色で表現してみる(色相を時間の関数とする)。matplotlibのscatter()に工夫を加える。この関数は点ごとに色を指定できるが、散布図向けの関数であるため、テント点の間の線画引けない。そこで連続した線分の代わりに補間のための中間点を配置する。

```python
k = 10 # 連続する2点の間に10の中間点を作る
# xとyをそれぞれ補間する
x2 = np.interp(np.arange(n*k), np.arange(n)*k, x)
y2 = np.interp(np.arange(n*k), np.arange(n)*k, y)
```

```python
# グラディエーション色を使って描画
plt.scatter(x2, y2, c=range(n*k), linewidths=0,
            marker='o', s=3, cmap=plt.cm.jet)
plt.axis('equal')
plt.xticks([]); plt.yticks([])
```

## 解説
- ブラウン運動W(t)にはいくつかの重要な性質がある。第一に連続的な奇跡を描く。第二に増分$W(t+\tau)-W(t)$は、重ならない時間感覚においてそれぞれ独立。そして第三に、その増分はガウス確率変数。より正確には次の式で表される。
\begin{equation}
\forall t, \tau > 0, W(t+\tau)-W(t) \sim N(0, \tau)
\end{equation}
- 特に、W(t)の密度は、分散tの正規分布となる。
- 加えて、ブラウン運動、より一般的に言う確率過程は偏微分方程式と密接な関係を持つ。拡散方程式の一種である熱方程式の基本解は、W(t)の密度となる。さらに一般化すると、確率微分方程式の解の密度により満たされる偏微分方程式の一種が**フォッカー・プランク方程式**であると言える。
